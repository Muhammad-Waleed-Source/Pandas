A **Pandas DataFrame** is a two-dimensional, tabular data structure with labeled rows(Series) and columns(Series)

```python
import numpy as np
import pandas as pd
```

### Creating DataFrame

```python
# Using lists (We need a 2D list for dataframe)

student_data = [
    [100, 80, 10,],
    [90, 70, 7],
    [120, 100, 14],
    [80, 50, 2]
]

pd.DataFrame(student_data, columns=['iq', 'marks', 'package'])
```

```python
# Using dicts (We do not need to specify col here)

student_dict = {
    "iq": [100, 90, 120, 80, 0, 0],
    'marks': [80, 70, 100, 50, 0, 0],
    'package': [10, 7, 14, 2, 0, 0]
}

students = pd.DataFrame(student_dict)
students
```

### Selecting rows from a DataFrame

- **iloc** - searches using index positions
- **loc** - searches using index labels

```python
student_dict = {
    "name": ['Waleed', 'Kamran', 'Faisal', 'Sabir', 'Hamza', 'Umair'],
    "iq": [100, 90, 120, 80, 0, 0],
    'marks': [80, 70, 100, 50, 0, 0],
    'package': [10, 7, 14, 2, 0, 0]
}

students_info = pd.DataFrame(student_dict)
students_info.set_index('name', inplace=True)

print(students_info)
```

```python
# single row
movies
```

```python
# iloc (Fetch based on index position)

print(movies.iloc[0])
```

```python
# multiple rows
movies.iloc[0:5]
```

```python
movies.iloc[5:16:2] # with difference of two
```

```python
# fancy indexing
movies.iloc[[0, 4, 5]]
```

```python
# loc (Fetch based on index label)
students_info
```

```python
students_info.loc['Waleed']
```

```python
students_info.loc['Waleed':"Sabir"]
```

```python
students_info.loc['Waleed':"Sabir":2]
```

```python
# fancy indexing
students_info.loc[['Faisal', 'Sabir', 'Waleed']]
```

```python
# We can also use iloc here, because Pandas has default indices always
students_info.iloc[0:5:2]
```
