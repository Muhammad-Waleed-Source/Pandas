# **1. Multi-index Series**

```python
# can we have multiple index? Let's try
index_val = [('cse',2019), ('cse', 2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]

a = pd.Series([1, 2, 3, 4, 5, 6, 7, 8], index=index_val)

print(a)
```

```python
print(a[('cse', 2021)])
```

```python
# The Problem? (What if we want to fetch all the values where the index is cse)

a['cse'] # this will give error
```

So the above method is not good to create a multi-index series.

```python
# The solution -> multiindex series(also known as Hierarchical Indexing)
# multiple index levels within a single index
```

```python
# how to create multiindex object

# 1. pd.MultiIndex.from_tuples()
index_val = [('cse',2019), ('cse', 2020),('cse',2021),('cse',2022),('ece',2019),('ece',2020),('ece',2021),('ece',2022)]

multiindex = pd.MultiIndex.from_tuples(index_val)
print(multiindex)
```

```python
# So pandas will decouple the indexes, let's see
print(multiindex.levels)
```

```python
# 2. pd.MultiIndex.from_product(). **Best way**

pd.MultiIndex.from_product([['cse', 'ece'], [2019, 2020, 2021, 2022]])
```

```python
# Creating a series with multi-index object
s = pd.Series([1,2,3,4,5,6,7,8], index=multiindex)
```

```python
# How to fetch items from such series
print(s['cse'])
```

```python
print(s['ece'])
```

```python
print(s[('cse', 2021)])
```

So multi-index series is 2D because we need two thing to fetch a specific record, whereas for a 1D series we need only one thing.

```python
# unstack (convertind a multi-index series to a dataframe)

temp = s.unstack()
print(temp)
```

```python
# Three ways to access items from a dataframe
print(temp.loc['cse', 2019])
print(temp.iloc[0, 0])
print(temp.at['cse', 2019])
```

```python
print(temp.loc['ece', 2019])
print(temp.iloc[1, 0])
print(temp.at['ece', 2019])
```

```python
# stack (Opposite of unstack, converts a dataframe into a multi-index series)

print(temp.stack()) # Got the original thing back
```

## **Question:**
Why we need a multi-index series if we can display the data using dataframe?

## **Answer:**
We can display a higher dimension data like 3D, 4D or 10D etc into a lower dimension data like 2D using two lower dimension objects called series and dataframe.

## Stacking and Unstacking

### Dataframe 1

```python
branch_df1
```

```python
print(branch_df1.unstack())

# As we have two levels of indexing in rows, program(L0) and year(L1) and also we have L0 indexing in col (avg_pacakge and students). So the L1(inner) indexing of row will become the L1 of col indexing when we perform unstack.
```

```python
print(branch_df1.unstack().unstack()) # so it is not a dataframe now, it is a series because it does not have rows now

# Also here the L0 indexing of row is converted to L2 of the col indexing
```

```python
# Now performing stack
print(branch_df1.unstack().stack())

# When we stack it, the opposite happens the col indexing converts into row indexing starting from L2.
```

```python
print(branch_df1.unstack().stack().stack())
```

## Sorting index

```python
branch_df3
```

```python
# as the indexes are already sorted so, sorting in decending order

branch_df3.sort_index(ascending=False)
```

```python
# What is we want to sort branch in decending but year in ascending

branch_df3.sort_index(ascending=[False, True])

# ascending=[level 0 index, level 1 index]
```

```python
# What if we want to sort at only one level (year -> level 1)

branch_df3.sort_index(level=1, ascending=False)
```

```python
# What if we want to sort at only one level (branch -> level 0)

branch_df3.sort_index(level=0, ascending=False)
```
