```python
import numpy as np
import pandas as pd
```

```python
# value_counts (work on series and dataframe)
a = pd.Series([1,1,1,2,2,3])
print(a.value_counts()) # Gives how much time a value is repeated in a series
```

```python
marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
],columns=['iq','marks','package'])

print(marks)
```

```python
print(marks.value_counts()) # Give how much time a row is repeated
```

```python
ipl = pd.read_csv('ipl-matches.csv')
ipl.head()
```

```python
# Find which player has won most Player of the Match in finals and qualifiers
potm = ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()

print(potm)
```
