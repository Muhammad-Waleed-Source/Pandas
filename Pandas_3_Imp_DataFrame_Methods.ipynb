# **value_counts**

```python
# value_counts (work on series and dataframe)
a = pd.Series([1,1,1,2,2,3])
print(a.value_counts()) # Gives how much time a value is repeated in a series
```

```python
marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
],columns=['iq','marks','package'])

print(marks)
```

```python
print(marks.value_counts()) # Give how much time a row is repeated
```

```python
ipl = pd.read_csv('ipl-matches.csv')
ipl.head()
```

```python
# Find which player has won most Player of the Match in finals and qualifiers
potm = ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()

print(potm)
```

# **sort_values**

```python
# Works on series as well as dataframe

print((ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending=False))
```

```python
x = pd.Series([12, 14, 1, 56, 89])
print(x.sort_values())
print(x.sort_values(ascending=False))
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
movies.sort_values('title_x', ascending=False)
```

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]

    }
)

print(students)
```

```python
# sort_values on dataframe with NaN values
print(students.sort_values('name')) # missing values are pushed to the end (default behavior)
```

```python
print(students.sort_values('name', na_position='first')) # Now the null values are pushed to the top
```
