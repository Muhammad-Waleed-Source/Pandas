## Working with multiple dataframes at once

```python
import numpy as np
import pandas as pd
```

```python
courses = pd.read_csv('/content/courses.csv')
deliveries = pd.read_csv('/content/deliveries.csv')
matches = pd.read_csv('/content/matches.csv')
nov = pd.read_csv('/content/reg-month1.csv')
dec = pd.read_csv('/content/reg-month2.csv')
students = pd.read_csv('/content/students.csv')
```

### **pd.concat:**
Used to combine multiple DataFrames or Series along a specific axis (either vertically as rows or horizontally as columns)

```python
# vertically
pd.concat([nov, dec])
```

```python
# There is an issue with the above concatenation that both the dataframe retained their index numbers.(0 to 24, 0 to 27). So to make it continous we will use ignore_index=True

regs = pd.concat([nov, dec], ignore_index=True)
```

```python
# multi-index col (For differentiation)
multi = pd.concat([nov, dec], keys=["Nov", 'Dec'])
multi
```

```python
multi.loc['Nov']
```

```python
multi.loc['Dec']
```

```python
print(multi.loc[('Nov', 1)]) # 2nd person
print(multi.loc[('Nov', 4)]) # 5th person
```

```python
# horizontally
pd.concat([nov, dec], axis=1)
```
