## Working with multiple dataframes at once

```python
import numpy as np
import pandas as pd
```

```python
courses = pd.read_csv('/content/courses.csv')
deliveries = pd.read_csv('/content/deliveries.csv')
matches = pd.read_csv('/content/matches.csv')
nov = pd.read_csv('/content/reg-month1.csv')
dec = pd.read_csv('/content/reg-month2.csv')
students = pd.read_csv('/content/students.csv')
```

### **pd.concat:**
Used to combine multiple DataFrames or Series along a specific axis (either vertically as rows or horizontally as columns)

```python
# vertically
pd.concat([nov, dec])
```

```python
# There is an issue with the above concatenation that both the dataframe retained their index numbers.(0 to 24, 0 to 27). So to make it continous we will use ignore_index=True

regs = pd.concat([nov, dec], ignore_index=True)
```

```python
# multi-index col (For differentiation)
multi = pd.concat([nov, dec], keys=["Nov", 'Dec'])
multi
```

```python
multi.loc['Nov']
```

```python
multi.loc['Dec']
```

```python
print(multi.loc[('Nov', 1)]) # 2nd person
print(multi.loc[('Nov', 4)]) # 5th person
```

```python
# horizontally
pd.concat([nov, dec], axis=1,keys=['Nov','Dec'])
```

### **Joinings:**
Combining two DataFrames based on shared columns or indexes, functioning exactly like a SQL JOIN. It aligns rows from a "left" and a "right" dataset using specified key columns to create a single, unified DataFrame.

### **1. Inner Join:**
It is a method used to combine two DataFrames based on a common key, returning only the rows that have matching values in both datasets. It acts like a mathematical intersection; if a key exists in one DataFrame but not the other, that row is entirely excluded from the final result.

```python
# Find the name of students enrolled in courses
students.merge(regs, how='inner', on='student_id')
```

### **2. Left Join:**
Keeps all rows from the left DataFrame and brings in matching rows from the right DataFrame based on a shared key. If a key in the left DataFrame does not exist in the right DataFrame, pandas fills the missing right-side columns with NaN (Not a Number) values

```python
courses.merge(regs, how='left', on='course_id')
```

### **3. Right Join:**
Keep all rows from the right (second) DataFrame and match them with rows from the left (first) DataFrame based on a common key. If a key exists in the right DataFrame but not in the left, the resulting columns from the left DataFrame will be filled with missing values (NaN).

```python
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students, temp_df], ignore_index=True)

students
```

```python
regs.tail()
```

```python
students
```

```python
students.merge(regs, how='right', on='student_id')
```

### **4. Outer/Full Join:**
Combines two DataFrames by returning all rows from both datasets, regardless of whether they have matching values in the join key.

```python
students.merge(regs, how='outer', on='student_id').tail(10)
```
