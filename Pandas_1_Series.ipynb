### Series using read_csv
The defaul behavior of read_csv is it reads the data as a Data Frame so we have to change it to a Series using squeeze('columns')

```python
# With one col
subs = pd.read_csv('subs.csv')
print(subs)
print(type(subs)) # the data is in Data Frame not Series
```

```python
# We have to use squeeze method in order to import data as a Series
subs = pd.read_csv('subs.csv').squeeze('columns')
# subs = pd.read_csv("subs.csv").iloc[:, 0] ..Same..
print(subs)
print(type(subs))
```

# **Why does movies[-1] work then?**

It's a historical behavior of pandas. For a Series with a non-integer index (like strings), series[-1] falls back to positional indexing. But for a Series with an integer index, -1 is treated as a label. This inconsistency is one reason pandas recommends always using .iloc for positions and .loc for labels. It makes your code unambiguous and future-proof.

# **Copy** and Views

The **copy()** function in pandas creates a duplicate of a DataFrame or a Series. It isolates your modifications, ensuring that changes made to the new object do not unexpectedly overwrite the original data.

```python
print(vk)
```

```python
new = vk.head()
print(new)
```

```python
new[1] = 100
# So we are changing the value in a new series, but it will also reflect in the original, so with the head() we get a view of our original data not a new copy of it (We got references to our original data)
```

```python
print(vk)
```

```python
# To avoid this we use copy() method
actual_new = vk.head().copy()
print(actual_new)
```

```python
actual_new[1] = 1
```

```python
print(actual_new)
```

```python
print(vk) # Now here the value is not changed
```
